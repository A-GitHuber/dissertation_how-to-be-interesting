```python
import pandas as pd
import matplotlib.pyplot as plt

# sheet
sheet_names = ['EXP', 'INT', 'difference', 'difference1', 'sign2', 'sign3']
dfs = [pd.read_excel('compare.xlsx', sheet_name=sheet_name) for sheet_name in sheet_names]

# 23
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# sheet
for ax, df, sheet_name in zip(axes.flatten(), dfs, sheet_names):
    ax.plot(df['Threshold'], df['Accuracy'], label='Accuracy')
    ax.plot(df['Threshold'], df['Precision'], label='Precision')
    ax.plot(df['Threshold'], df['Recall'], label='Recall')
    ax.plot(df['Threshold'], df['F1'], label='F1')
    ax.set_title(sheet_name)
    ax.legend()

# 
plt.tight_layout()

# 
plt.savefig('metric.pdf', format='pdf')
plt.show()
```
